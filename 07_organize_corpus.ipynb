```python
import os
import traceback
from shutil import copyfile
```

```python
corpus_path = "./corpus"

os.makedirs(corpus_path, exist_ok=True)
os.makedirs(f"{corpus_path}/equal", exist_ok=True)
os.makedirs(f"{corpus_path}/genderless", exist_ok=True)
os.makedirs(f"{corpus_path}/man", exist_ok=True)
os.makedirs(f"{corpus_path}/woman", exist_ok=True)
```

```python
exclude_folders = [".ipynb_checkpoints"]
```

```python
articles_path = "./articles"
articles_folders = [
    dir 
    for dir in os.listdir(articles_path) 
    if not os.path.isfile(f"{articles_path}/{dir}")
    and not dir in exclude_folders
]
```

```python
for dir in articles_folders:
    dir_path = f"{articles_path}/{dir}"
        
    try:
        copyfile(f"{dir_path}/equal.txt", f"{corpus_path}/equal/{dir}_equal.txt")
        copyfile(f"{dir_path}/genderless.txt", f"{corpus_path}/genderless/{dir}_genderless.txt")
        copyfile(f"{dir_path}/man.txt", f"{corpus_path}/man/{dir}_man.txt")
        copyfile(f"{dir_path}/woman.txt", f"{corpus_path}/woman/{dir}_woman.txt")
    except Exception as e:
        print("Error occured when trying to organize corpus for", dir_path)
        traceback.print_exc()
```
