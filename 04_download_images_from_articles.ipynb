```python
import os
import pandas as pd
from typing import Iterable
from helpers import get_request, get_image_name, get_page_name
```

```python
filename = "./article_links_16-09-25_12-21-53_with_contents_with_data"
articles_df = pd.read_csv(f"{filename}.csv")
articles_df
```

```python
def download_images(links: Iterable[str], images_srcs: Iterable[str]):
    def download_images_on_page(page_link: str, image_srcs: str):
        try:
            image_srcs = image_srcs.split("\n")

            folder = f"images/{get_page_name(page_link)}"
            if os.path.exists(folder):
                return
            os.makedirs(folder)
            
            for src in image_srcs:
                filename = get_image_name(src)

                with open(f"{folder}/{filename}", "wb") as file:
                    image_content = get_request(src).content
                    file.write(image_content)
        except Exception as e:
            print("Error downloading image", src, "from", page_link)
            print(e)
    
    zipped = zip(links, images_srcs)
    list_len = len(links)

    for (i, (link, image_srcs)) in enumerate(zipped):
        if i % 10 == 0:
            print("Getting images from page", i, "/", list_len)

        download_images_on_page(link, image_srcs)
```

```python
download_images(articles_df["link"], articles_df["images_srcs"])
```
