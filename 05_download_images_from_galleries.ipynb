```python
import os
import pandas as pd
from bs4 import BeautifulSoup
from typing import Iterable
from helpers import get_request, get_image_name, get_page_name
```

```python
links_df = pd.read_csv("./article_links_16-09-25_12-21-53.csv")
links_df
```

```python
gallery_links = links_df[links_df["link"].str.contains("gallery")]
gallery_links
```

```python
def download_images(links: Iterable[str]):
    def download_images_on_page(page_link: str):
        try:
            page_content = get_request(f"https://ufc.com{page_link}").content

            soup = BeautifulSoup(page_content, "html.parser")
            images = soup.find_all("img")
            image_srcs = [image.get("src") for image in images]

            folder = f"images/{get_page_name(page_link)}"
            if os.path.exists(folder):
                return

            os.makedirs(folder)
            
            for src in image_srcs:
                filename = get_image_name(src)

                with open(f"{folder}/{filename}", "wb") as file:
                    image_content = get_request(src).content
                    file.write(image_content)
        except Exception as e:
            print("Error downloading image", src, "from", page_link)
            print(e)
    
    list_len = len(links)

    for (i, link) in enumerate(links):
        print("Getting images from page", i, "/", list_len)
        download_images_on_page(link)
```

```python
download_images(gallery_links["link"])
```
